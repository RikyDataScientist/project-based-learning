# Import Dependency

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Plot Customization

```python
x = np.array([2020, 2021, 2022, 2023])
y1 = np.array([100, 165, 200, 175])
y2 = np.array([90, 180, 130, 160])
y3 = np.array([80, 150, 170, 190])

style = dict(
    marker='.', markersize=20, markerfacecolor='#1CD3FC', markeredgecolor="#176272",
    linestyle='dashed', linewidth='2'
)

plt.plot(
    x, y1,
    **style
    )

plt.plot(
    x, y2,
    **style
)

plt.title('Yearly Data')
plt.xlabel('Year')
plt.ylabel('Value')

plt.show()
```

# Labels & Grid Lines

```python
plt.plot(
    x, y1
)

plt.plot(
    x, y2
)

plt.plot(
    x, y3
)

sty = dict(
    fontweight='bold',
    family='Arial'
)

plt.grid(color='lightgray')

plt.title('Yearly Data', size=16, **sty)
plt.xlabel('Year', size=12, **sty)
plt.ylabel('Value', size=12, **sty)

plt.xticks(x)

plt.show()
```

```python
# Bar & Pie Chart
categories = np.array(['Grains', 'Fruit', 'Vegetables', 'Protein', 'Dairy', 'Sweets'])
value = np.array([4, 3, 2, 5, 6, 3])
colors = [
    "#457B9D",
    "#A8DADC",
    "#B8A1E3",
    "#E9C46A",
    "#F4A261",
    "#E76F51"
]

def bar_pie_chart(cat, val):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

    plt.suptitle('Daily Consumption', fontsize=24, fontweight="550", y=0.98)

    ax1.bar(cat, val, width=0.8, color=colors, zorder=3)

    ax1.set_ylabel('Quantity', size=12)

    ax1.grid(axis='y', linestyle='--', alpha=0.3, zorder=0)
    ax1.spines['top'].set_visible(False)
    ax1.spines['right'].set_visible(False)

    ax2.pie(val, labels=cat, autopct="%1.1f%%", colors=colors)

    plt.tight_layout()

    plt.show()

# Histogram & BoxPlot
data = np.random.normal(loc=80, scale=6, size=1000)
scores = [65, 70, 72, 75, 80, 82, 85, 88, 90, 95, 150]

def histogram(arr):
    fig, ax = plt.subplots(figsize=(8, 5))
    plt.suptitle('Student Score', fontsize=24, fontweight='550', y=0.98)

    ax.hist(arr, bins=30, alpha=0.8, edgecolor='black', zorder=3)

    ax.grid(axis='y', linestyle='--', alpha=0.1, zorder=0)

    ax.set_xlabel('Score', size=12)
    ax.set_ylabel('Frequency', size=12)

    ax.spines[['top', 'right']].set_visible(False)

    plt.tight_layout()

    plt.show()

def boxplot(arr):
    fig, ax = plt.subplots(figsize=(8, 5))
    plt.suptitle('Student Score Distribution', fontsize=24, fontweight='550', y=0.98)

    ax.boxplot(arr, zorder=3)
    ax.grid(axis='y', linestyle='--', alpha=0.1, zorder=0)

    ax.spines[['top', 'right']].set_visible(False)

    ax.set_ylabel('Score', size=12)

    plt.tight_layout()

    plt.show()
    

if __name__ == "__main__":
    bar_pie_chart(categories, value)
    histogram(data)
    boxplot(scores)
```
