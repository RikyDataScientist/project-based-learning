# Import Depedency

```python
import pandas as pd
import numpy as np
```

# Series

Series is a one-dimensional labeled array capable of holding any data type (integers, strings, floating point numbers, Python objects, etc.). The axis labels are collectively referred to as the index.

```python
data = [12, 15, 14, 10, 8]
s = pd.Series(data, index=['a', 'b', 'c', 'd', 'e'])

s.loc['c'] = 90

print(s, '\n')
print(s.iloc[1], '\n')
print(s.loc['a':'c'], '\n')
print(s[s > 12])
```

# Data Frame

DataFrame is a 2-dimensional labeled data structure with columns of potentially different types. It is similar to a spreadsheet or SQL table, or a dict of Series objects. It is generally the most commonly used pandas object. Like Series, DataFrame accepts many different kinds of input:

- Dict of 1D ndarrays, lists, dicts, or Series
- 2-D numpy.ndarray

```python
employess = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'Department': ['HR', 'IT', 'Finance', 'Marketing']
}

df1 = pd.DataFrame(employess)

# Add New Column
df1['Gender'] = ['Female', 'Male', 'Male', 'Male']

# Add New Row
new_employee = [
    {'Name': 'Eve', 'Age': 28, 'Department': 'IT', 'Gender': 'Female'},
    {'Name': 'Ais', 'Age': 24, 'Department': 'Sales', 'Gender': 'Female'}
    ]
df1 = pd.concat([df1, pd.DataFrame(new_employee)], ignore_index=True)

print(df1, '\n')
print(df1.iloc[2], '\n')
```

# Importing

```python
dataset = '../../00-datasets/source/data.csv'

df = pd.read_csv(dataset)

print(df, '\n')
print(df.to_string())
```

# Selection

```python
print(df.head(5))

# Selection By Column
print(df['Age'].head(5), '\n')
print(df['Country'].head(5), '\n')
print(df[['Salary', 'Purchased', 'Age']].head(5), '\n')

# Selection By Row
print(df.loc[0], '\n')
print(df.loc[2:4, ['Age', 'Country']], '\n')

print(df.iloc[1:3, 0:3], '\n')
```

# Filtering

```python
df2 = pd.read_csv('../../00-datasets/source/gammafest/train.csv')

print(df2.head(5), '\n')

friendly = df2[df2['tournament'] == 'Friendly']
print(friendly[['Id', 'match_id']].head(5), '\n')

scotland = df2[df2['team'] == 'Scotland']
print(scotland[['Id', 'match_id']].head(5), '\n')

scotland_match = df2[(df2['team'] == 'Scotland') | (df2['opponent'] == 'Scotland')]
print(scotland_match[['Id', 'match_id']].head(5), '\n')
```

# Aggregate

```python
# Whole Column
print(df.nunique(), '\n')
print(df.count(numeric_only=True), '\n')
print(df.mean(numeric_only=True), '\n')
print(df.sum(numeric_only=True), '\n')
print(df.min(numeric_only=True), '\n')
print(df.max(numeric_only=True), '\n')
print(df.std(numeric_only=True), '\n')

# Specific Column
print(df['Salary'].count(), '\n')
print(df['Salary'].mean(), '\n')
print(df['Salary'].sum(), '\n')
print(df['Age'].min(), '\n')
print(df['Age'].max(), '\n')
print(df['Salary'].std(), '\n')

# Group
salary_by_country = df.groupby('Country')['Salary'].mean()
print(salary_by_country, '\n')

age_by_country = df.groupby('Country')['Age']
print(age_by_country.min(), '\n')
print(age_by_country.max(), '\n')
```

# Data Cleaning

```python
print(df.info(), '\n')

# Drop Irrelevant Columns
df1 = df.drop(columns=['Age'])
print(df1.head(), '\n')

data = {
    'Country': ['USA', 'canada', 'mexico'],
    'Age': [25, 30, np.nan],
    'Salary': [50000, 60000, 55000],
    'Purchased': ['Yes', np.nan, 'No']
}

df = pd.concat([df, pd.DataFrame(data)], ignore_index=True)
print(df, '\n')

# Handle Missing Values
df = df.dropna(subset=['Country'])
print(df, '\n')

df = df.fillna({
    'Age': df['Age'].mean(),
    'Purchased': 'None',
    'Salary': df['Salary'].mean()
})
print(df, '\n')

# Fix Inconsistent Data
df['Country'] = df['Country'].replace({
    'USA': 'United States',
    'canada': 'Canada',
    'mexico': 'Mexico'
})

# Standardize Data
df['Country'] = df['Country'].str.title()

# Fix Data Types
df['Adult'] = df['Age'] >= 18
df['Adult'] = df['Adult'].astype(bool)

# Remove Duplicates
df = df.drop_duplicates()

print(df)
```
